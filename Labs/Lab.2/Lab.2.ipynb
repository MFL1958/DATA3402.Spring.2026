*Exercise 4:* Write a function that takes a `n` by `n` matrix representing a tic-tac-toe game, and returns -1, 0, 1, or 2 indicating the game is incomplete, the game is a draw, player 1 has won, or player 2 has one, respectively. Here are some example inputs you can use to test your code:

````python
# Write your solution here
def check_game_state(board):
    n = len(board)
    # we check each row to see if they all have the same integer (1 or 2) then 
    # return that specific integer as the winner
    for row in board:
        if row.count(row[0]) == n and row[0] != 0:
            return row[0]
    for col in range(n):
        column = [board[row][col] for row in range(n)]
        if column.count(column[0]) == n and column[0] != 0:
            return column[0]
    # diagonal win conditions:
    diag1 = [board[i][i] for i in range(n)]
    if diag1.count(diag1[0]) == n and diag1[0] != 0:
        return diag1[0]
    diag2 = [board[i][n-1-i] for i in range(n)]
    if diag2.count(diag2[0]) == n and diag2[0] != 0:
        return diag2[0]

    for row in board:
        if 0 in row:
            return -1
     # draw if none of the options are satisfied       
    return 0
````

*Exercise 5:* Write a function that takes a game board, player number, and `(x,y)` coordinates and places "X" or "O" in the correct location of the game board. Make sure that you only allow filling previously empty locations. Return `True` or `False` to indicate successful placement of "X" or "O".

````python
# Write your solution here
def valid_move(board, player, x, y):
    # first we check for a valid placement in the game grid
    if x < 0 or x >= len(board) or y < 0 or y >= len(board):
        return False
    # check if empty
    if board[x][y] != 0:
        return False
    board[x][y] = player
    return True
````

````python
# Test your solution here
board = create_board(3)

print(valid_move(board, 1, 0, 0))  # True (X placed)
print(valid_move(board, 2, 0, 0))  # False (already occupied)
print(valid_move(board, 2, 1, 1))  # True (O placed)

draw_board(board)
````

*Exercise 7:* Write a function that takes a board, player number, and location specified as in exercise 6 and then calls exercise 5 to correctly modify the board.  

````python
# Write your solution here
def place_move(board, player, location):
    letter_location = location[0].upper()
    col = ord(letter_location) - ord("A")
    row = int(location[1]) - 1
    
    return valid_move(board, player, row, col)
````

*Exercise 8:* Write a function is called with a board and player number, takes input from the player using python's `input`, and modifies the board using your function from exercise 7. Note that you should keep asking for input until you have gotten a valid input that results in a valid move.

````python
# Write your solution here
def get_player_move(board, player):
    while True:
        location = input("Player " + str(player) + ", enter your move (A1, B2, etc.): \n")
        if place_move(board, player, location):
            return
        else:
            print("Invalid move. Try again.")
````

*Exercise 9:* Use all of the previous exercises to implement a full tic-tac-toe game, where an appropriate board is drawn, 2 players are repeatedly asked for a location coordinates of where they wish to place a mark, and the game status is checked until a player wins or a draw occurs.

````python
# Write your solution here
def TicTacToe(n):
    board = create_board(n)
    player_turn = 1
    while check_game_state(board) == -1:
        draw_board(board)
        get_player_move(board, player_turn)
        player_turn = 2 if player_turn == 1 else 1

    draw_board(board)
    winner = check_game_state(board)
    return "Draw." if winner == 0 else f"Player {winner} Has Won."
````

````python
# Write you solution here
def computer_move(board, computer):
    human = 1 # if computer == 2 else 2
    move = computer_win_check(board, computer)
    if move is not None:
        valid_move(board, computer, move[0], move[1])
        return
    human_win_check = computer_win_check(board, human)
    if human_win_check is not None:
        valid_move(board, computer, human_win_check[0], human_win_check[1])
        return
    for i in range(len(board)):
        for j in range(len(board)):
            if board[i][j] == 0:
                valid_move(board, computer, i, j)
                return
    
def computer_win_check(board, player):
    n = len(board)
    for i in range(n):
        for j in range(n):
            if board[i][j] == 0:
                board[i][j] = player
                if check_game_state(board) == player:
                    board[i][j] = 0
                    return (i, j)
                board[i][j] = 0
    return None

def ComputerTicTacToe(n):
    board = create_board(n)
    player_turn = 1
    while check_game_state(board) == -1:
        draw_board(board)
        if player_turn == 1:
            get_player_move(board, 1)
        else:
            computer_move(board, 2)
        player_turn = 2 if player_turn == 1 else 1
    draw_board(board)
    winner = check_game_state(board)
    return "Draw." if winner == 0 else (f"Player {winner} Has Won." if winner == 1 else "Computer Wins")
    
````
